```python
import matplotlib
# matplotlib.use('Agg')
import matplotlib.pyplot as plt
# plt.rc('text', usetex=True)
from plot import Plotter
import math
```

```python
def plot_mujoco():
    plotter = Plotter()
    games = [
        # 'Ant-v2',
        'HalfCheetah-v2',
        'Walker2d-v2',
        'Hopper-v2',
        # 'Swimmer-v2',
        # 'Reacher-v2',
    ]

    patterns = [
        'correction-GradientDICE',
        'correction-None',
        # 'correction-GradientDICE_discount-1.0',
        # 'correction-None_discount-1.0',
        # 'expert'
    ]

    labels = patterns

    plotter.plot_games(games=games,
                       patterns=patterns,
                       agg='mean',
                       downsample=0,
                       labels=labels,
                       right_align=True,
                       tag=plotter.RETURN_TEST,
                       root='./20200628',
                       interpolation=0,
                       window=4,
                       right_most=0,
                       )

    plt.tight_layout()
    plt.savefig('mujoco_eval.svg', bbox_inches='tight')
```

```python
if __name__ == '__main__':
    import os
    import tensorflow as tf
    tf.logging.set_verbosity(tf.logging.ERROR) # to compress verbose info
    plot_mujoco()
    plt.show()
```
